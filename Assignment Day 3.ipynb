Question 2
Using for loop please print all the prime numbers between 1- 200 using FOR LOOP AND RANGE
function.
Output2,3,5,7…..

```python
for Number in range (1, 200):
    count = 0
    for i in range(2, (Number//2 + 1)):
        if(Number % i == 0):
            count = count + 1
            break

    if (count == 0 and Number != 1):
        print(" %d" %Number, end = '  ')
```
